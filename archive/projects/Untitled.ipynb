```python
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart

def send_email(subject, message, to_email, df):
    # Gmail account credentials
    gmail_user = 'bobbys.daily.lines@gmail.com'  # Your email
    gmail_app_password = 'zytx hudf gwky pdht'  # Your app password

    # Convert DataFrame to HTML
    tyled_df = df.style.set_properties(**{
        'background-color': 'white',
        'color': 'black',
        'border-color': 'grey',
        'border-style': 'solid',
        'border-width': '1px',
        'text-align': 'left'
    }).set_table_styles([
        {'selector': 'th',
         'props': [('background-color', '#007bff'),
                   ('color', 'white'),
                   ('border-color', 'grey')]}
    ]).render()

    html = f"""\
    <html>
      <head>
        <style>
          body {{
            font-family: Arial, sans-serif;
            margin: 0;
            padding: 0;
            background-color: #f4f4f4;
            color: #333;
          }}
          h1 {{
            background-color: #007bff;
            color: white;
            padding: 10px;
          }}
          p {{
            padding: 5px 10px;
          }}
          .dataframe {{
            border-collapse: collapse;
            width: 100%;
            margin: 10px 0;
          }}
          .dataframe th, .dataframe td {{
            border: 1px solid #ddd;
            padding: 8px;
            text-align: left;
          }}
          .dataframe tr:nth-child(even){{background-color: #f2f2f2;}}
          .dataframe tr:hover {{background-color: #ddd;}}
        </style>
      </head>
      <body>
        <h1>Good Morning</h1>
        <p>Good Morning Mr. Plourde,</p>
        <p>Below are your daily picks for today's lines. Have a great day!!</p>
        {styled_df}
      </body>
    </html>
    """
    return html

    # Email content setup
    msg = MIMEMultipart()
    msg['From'] = gmail_user
    msg['To'] = to_email
    msg['Subject'] = subject

    msg.attach(MIMEText(html, 'html'))

    # Sending email
    try:
        server = smtplib.SMTP('smtp.gmail.com', 587)
        server.starttls()
        server.login(gmail_user, gmail_app_password)
        server.sendmail(gmail_user, to_email, msg.as_string())
        server.quit()
        print("Email sent successfully!")
    except Exception as e:
        print(f"Error: {e}")
# Usage
send_email('Test Subject', 'This is a test message', 'bobby.plourde12@yahoo.com')
```
